# Comparison with newman-triangular random graph.


3-clique is the same as triangular clustering. Therefore, if the script is correct, then the result of clique random graph with only $3$-cliques should show similar result with random triangular random model.


```python
import pickle
import matplotlib
import matplotlib.pyplot as plt

list_of_edge_triangle_ratio = [i*0.02 for i in range(51)]

data_compilation = pickle.load(open('./comparison.txt', 'rb'))
clique_triangle_clustering = data_compilation['random_clique']['triangle_clustering']
clique_average_shortest_patlen = data_compilation['random_clique']['average_shortest_patlen']
clique_list_of_mean_number_of_cliques = data_compilation['random_clique']['number_of_cliques']

tri_triangle_clustering = data_compilation['random_triangle']['triangle_clustering']
tri_average_shortest_patlen = data_compilation['random_triangle']['average_shortest_patlen']
tri_list_of_mean_number_of_cliques = data_compilation['random_triangle']['number_of_cliques']


%matplotlib inline
plt.figure(figsize = (15,10))
plt.subplot(221)
plt.plot([i*0.02 for i in range(51)], clique_triangle_clustering, 'rs', markersize = 5, label = 'Clique')
plt.plot([i*0.02 for i in range(51)], tri_triangle_clustering, 'bs', markersize = 5, label = 'Triangle')
font={'family' : 'normal', 'weight' : 'bold', 'size' : 15}
matplotlib.rc('font', **font)
plt.legend(loc=2)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Triangle clustering')

plt.subplot(222)
plt.plot([i*0.02 for i in range(51)], clique_average_shortest_patlen, 'rs', markersize = 5, label = 'Clique')
plt.plot([i*0.02 for i in range(51)], tri_average_shortest_patlen, 'bs', markersize = 5, label = 'Triangle')
plt.legend(loc=2)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Average shortest path length')

plt.subplot(223)
plt.plot([i*0.02 for i in range(51)], clique_list_of_mean_number_of_cliques, 'rs', markersize = 5, label = 'Clique')
plt.plot([i*0.02 for i in range(51)], tri_list_of_mean_number_of_cliques, 'bs', markersize = 5, label = 'Triangle')
plt.legend(loc=2)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Number of cliques')
```
