## 1. Edge triangle ratio vs triangular clustering coefficient

The mechanism of `newman_clustering_configuration` is as follows:



Here we investigate the increase of triangular clustering coefficient and the increase of the number of cliques in networks with the increse of edge-clustering ratio.

### CLUSTERING

```python
%matplotlib inline
import sys, os
sys.path.append(os.path.abspath('../../../'))
import functions.distribution_functions.poisson_dist.poisson as pd
import networkx as nx

import newman_triangle as nt
import numpy as np
        
list_of_mean_clustering_coefficient = []
list_of_mean_number_of_cliques = []
list_of_edge_triangle_ratio = [i*0.02 for i in range(51)]

for edge_triangle_ratio in list_of_edge_triangle_ratio:
    list_of_clustering_coefficient = []
    list_of_number_of_cliques = []
    for i in range(100):
        deg_seq = [pd.poisson_random_variable(5) for i in range(100)]
        G = nt.newman_clustering_configuration(deg_seq, edge_triangle_ratio)
        list_of_clustering_coefficient.append(np.mean(nx.clustering(G).values()))
        list_of_number_of_cliques.append(len([i for i in nx.find_cliques(G) if len(i) >= 4]))
    list_of_mean_number_of_cliques.append(np.mean(list_of_number_of_cliques))
    list_of_mean_clustering_coefficient.append(np.mean(list_of_clustering_coefficient))
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

plt.figure(figsize = (10, 7))
plt.plot(list_of_edge_triangle_ratio, list_of_mean_clustering_coefficient, 'rs', markersize = 15)
font={'family' : 'normal', 'weight' : 'bold', 'size' : 22}
matplotlib.rc('font', **font)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Triangle clustering')
plt.show()
```
