## 1. Edge square ratio vs triangular clustering coefficient

The mechanism of `newman_clustering_configuration` is as follows:



Here we investigate the increase of triangular clustering coefficient and the increase of the number of cliques in networks with the increse of edge-clustering ratio.

## 1. Triangle

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import pickle

data_compilation = pickle.load(open('properties_triangle.txt', 'rb'))
list_of_clique_ratio = [i*0.02 for i in range(51)]

plt.figure(figsize = (15, 10))
plt.subplot(221)
plt.plot(list_of_clique_ratio, data_compilation['triangular_clustering'], 'rs', markersize = 15)
font={'family' : 'normal', 'weight' : 'bold', 'size' : 15}
matplotlib.rc('font', **font)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Triangle clustering')

plt.subplot(222)
plt.plot(list_of_clique_ratio, data_compilation['square_clustering'], 'rs', markersize = 15)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Square clustering')

plt.subplot(223)
plt.plot(list_of_clique_ratio, data_compilation['average_shortest_patlen'], 'rs', markersize = 15)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Average shortest path length')

plt.subplot(224)
plt.plot(list_of_clique_ratio, data_compilation['number_of_cliques'], 'rs', markersize = 15)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Number of cliques')
```

## 4. If the triangle_ratio, square_ratio, and clique_ratio are all zero

Then the characteristic of the networks should be identical.

```python
%matplotlib inline
import pickle
import matplotlib.pyplot as plt
import matplotlib

list_of_triangular_clustering_coefficient_tri = []
list_of_square_clustering_coefficient_tri = []
list_of_average_shortest_patlen_tri = []
list_of_number_of_cliques_tri = []
list_of_degree_distribution_tri = []

list_of_triangular_clustering_coefficient_sqr = []
list_of_square_clustering_coefficient_sqr = []
list_of_average_shortest_patlen_sqr = []
list_of_number_of_cliques_sqr = []
list_of_degree_distribution_sqr = []

list_of_triangular_clustering_coefficient_cliq_4 = []
list_of_square_clustering_coefficient_cliq_4 = []
list_of_average_shortest_patlen_cliq_4 = []
list_of_number_of_cliques_cliq_4 = []
list_of_degree_distribution_cliq_4 = []

list_of_triangular_clustering_coefficient_cliq_5 = []
list_of_square_clustering_coefficient_cliq_5 = []
list_of_average_shortest_patlen_cliq_5 = []
list_of_number_of_cliques_cliq_5 = []
list_of_degree_distribution_cliq_5 = []

list_of_triangular_clustering_coefficient_cliq_6 = []
list_of_square_clustering_coefficient_cliq_6 = []
list_of_average_shortest_patlen_cliq_6 = []
list_of_number_of_cliques_cliq_6 = []
list_of_degree_distribution_cliq_6 = []


for realization in range(100):
    data_compilation = pickle.load(open('./net_Str_comparison_data/netstr_comparison' + str(realization) + '.txt', 'rb'))
    
    list_of_triangular_clustering_coefficient_tri.append(data_compilation['triangle']['triangular_clustering'])
    list_of_square_clustering_coefficient_tri.append(data_compilation['triangle']['square_clustering'])
    list_of_average_shortest_patlen_tri.append(data_compilation['triangle']['average_shortest_patlen'])
    list_of_number_of_cliques_tri.append(data_compilation['triangle']['number_of_cliques'])
    list_of_degree_distribution_tri.append(data_compilation['triangle']['degree_distribution'])

    list_of_triangular_clustering_coefficient_sqr.append(data_compilation['square']['triangular_clustering'])
    list_of_square_clustering_coefficient_sqr.append(data_compilation['square']['square_clustering'])
    list_of_average_shortest_patlen_sqr.append(data_compilation['square']['average_shortest_patlen'])
    list_of_number_of_cliques_sqr.append(data_compilation['square']['number_of_cliques'])
    list_of_degree_distribution_sqr.append(data_compilation['square']['degree_distribution'])

    list_of_triangular_clustering_coefficient_cliq_4.append(data_compilation['clique_4']['triangular_clustering'])
    list_of_square_clustering_coefficient_cliq_4.append(data_compilation['clique_4']['square_clustering'])
    list_of_average_shortest_patlen_cliq_4.append(data_compilation['clique_4']['average_shortest_patlen'])
    list_of_number_of_cliques_cliq_4.append(data_compilation['clique_4']['number_of_cliques'])
    list_of_degree_distribution_cliq_4.append(data_compilation['clique_4']['degree_distribution'])

    list_of_triangular_clustering_coefficient_cliq_5.append(data_compilation['clique_5']['triangular_clustering'])
    list_of_square_clustering_coefficient_cliq_5.append(data_compilation['clique_5']['square_clustering'])
    list_of_average_shortest_patlen_cliq_5.append(data_compilation['clique_5']['average_shortest_patlen'])
    list_of_number_of_cliques_cliq_5.append(data_compilation['clique_5']['number_of_cliques'])
    list_of_degree_distribution_cliq_5.append(data_compilation['clique_5']['degree_distribution'])

    list_of_triangular_clustering_coefficient_cliq_6.append(data_compilation['clique_6']['triangular_clustering'])
    list_of_square_clustering_coefficient_cliq_6.append(data_compilation['clique_6']['square_clustering'])
    list_of_average_shortest_patlen_cliq_6.append(data_compilation['clique_6']['average_shortest_patlen'])
    list_of_number_of_cliques_cliq_6.append(data_compilation['clique_6']['number_of_cliques'])
    list_of_degree_distribution_cliq_6.append(data_compilation['clique_6']['degree_distribution'])
    

    
    
    
    
plt.figure(figsize = (15, 10))
plt.subplot(221)
plt.plot(list_of_triangular_clustering_coefficient_tri, marker = 's', color = '#e41a1c', markersize = 15)
plt.plot(list_of_triangular_clustering_coefficient_sqr, marker = 's', color = '#377eb8', markersize = 15)
plt.plot(list_of_triangular_clustering_coefficient_cliq_4, marker = 's', color = '#4daf4a', markersize = 15)
plt.plot(list_of_triangular_clustering_coefficient_cliq_5, marker = 's', color = '#984ea3', markersize = 15)
plt.plot(list_of_triangular_clustering_coefficient_cliq_6, marker = 's', color = '#ff7f00', markersize = 15)
font={'family' : 'normal', 'weight' : 'bold', 'size' : 15}
matplotlib.rc('font', **font)
plt.xlabel('Edge ratio')
plt.ylabel('Triangle clustering')


plt.subplot(222)
plt.plot(list_of_square_clustering_coefficient_tri, marker = 's', color = '#e41a1c', markersize = 15)
plt.plot(list_of_square_clustering_coefficient_sqr, marker = 's', color = '#377eb8', markersize = 15)
plt.plot(list_of_square_clustering_coefficient_cliq_4, marker = 's', color = '#4daf4a', markersize = 15)
plt.plot(list_of_square_clustering_coefficient_cliq_5, marker = 's', color = '#984ea3', markersize = 15)
plt.plot(list_of_square_clustering_coefficient_cliq_6, marker = 's', color = '#ff7f00', markersize = 15)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Square clustering')

plt.subplot(223)
plt.plot(list_of_average_shortest_patlen_tri, marker = 's', color = '#e41a1c', markersize = 15)
plt.plot(list_of_average_shortest_patlen_sqr, marker = 's', color = '#377eb8', markersize = 15)
plt.plot(list_of_average_shortest_patlen_cliq_4, marker = 's', color = '#4daf4a', markersize = 15)
plt.plot(list_of_average_shortest_patlen_cliq_5, marker = 's', color = '#984ea3', markersize = 15)
plt.plot(list_of_average_shortest_patlen_cliq_6, marker = 's', color = '#ff7f00', markersize = 15)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Average shortest path length')

plt.subplot(224)
plt.plot(list_of_number_of_cliques_tri, marker = 's', color = '#e41a1c', markersize = 15, alpha = 0.3)
plt.plot(list_of_number_of_cliques_sqr, marker = 's', color = '#377eb8', markersize = 15, alpha = 0.3)
plt.plot(list_of_number_of_cliques_cliq_4, marker = 's', color = '#4daf4a', markersize = 15, alpha = 0.3)
plt.plot(list_of_number_of_cliques_cliq_5, marker = 's', color = '#984ea3', markersize = 15, alpha = 0.3)
plt.plot(list_of_number_of_cliques_cliq_6, marker = 's', color = '#ff7f00', markersize = 15, alpha = 0.3)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Number of cliques')

plt.figure(figsize = (15, 10))
plt.hist(list_of_degree_distribution_tri, alpha = 0.3)
plt.hist(list_of_degree_distribution_sqr, alpha = 0.3)
plt.hist(list_of_degree_distribution_cliq_4, alpha = 0.3)
plt.hist(list_of_degree_distribution_cliq_5, alpha = 0.3)
plt.hist(list_of_degree_distribution_cliq_6, alpha = 0.3)
plt.xlabel('Edge clustering ratio')
plt.ylabel('Degree distribution')
```
