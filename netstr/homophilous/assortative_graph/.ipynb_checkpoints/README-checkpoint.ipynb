# Assortative network

Through this notebook, we'll go through three different ways of creating networks with assortativity:
1. Using mixing algorithm used in Newman's paper: ["Assortative mixing in networkx"](http://arxiv.org/pdf/cond-mat/0205405v1.pdf)<br />
=> Rewire edges with probability calculated based on desired assortative / disassortative edge fraction
2. and another paper of Newman's: ["Mixing patterns in networks"](http://arxiv.org/pdf/cond-mat/0209450.pdf) <br />
=> Create a network with degree sequence created using given assortativity / disassortativity
3. Kristina Lerman's method: ["The Majority Illusion in Social Networks"](http://arxiv.org/pdf/1506.03022.pdf) <br />
=> Rewire edges until the desired assortativity is obtained

<br />
This instruction is composed of two sections:
1. Calculating assortatively numerically / analytically
2. Create a random graph with given assortativity, by <br/>
    2.1. Rewiring, with certain probability, to the limit <br/>
    2.2. Configuration model <br/>
    2.3. Rewire until assortativity exceeds desired assortativity.

## 1. Calculate assortativity of a graph numerically and analytically.

### 1.1 Investigate assortativity of a network

First, let's try calculating assortativity of a network, with half of nodes having timer $\tau=0$ and the rest half has timer $8$.

```python
import networkx as nx

G = nx.fast_gnp_random_graph(1000, .005)

for i in G:
    if i < 50:
        G.node[i]['timer'] = 0
    else:
        G.node[i]['timer'] = 8
        
print nx.attribute_assortativity_coefficient(G,'timer')
```

However, the result you obtained as a result of above calculation is not accurate, as it is just one realization of a network. Therefore, we should calculate __the average attribute assortativity over ensemble__ of ER graph, which should give $\sim0$ as ER graph is homogeneous graph.

```python
list_of_attribute_assortativity_coefficient = []
for i in xrange(1000):
    G = nx.fast_gnp_random_graph(1000, .005)

    for node_idx in G:
        if node_idx < 500:
            G.node[node_idx]['timer'] = 0
        else:
            G.node[node_idx]['timer'] = 8
    list_of_attribute_assortativity_coefficient.append(nx.attribute_assortativity_coefficient(G,'timer'))
    
import numpy as np
print np.mean(list_of_attribute_assortativity_coefficient)
```

We can observe the histogram of assortative and disassortative edges in a sample network (which is automaticallly the last network in the ensemble).

```python
edge_size = G.size()
# Edge distribution
# 1: disassortative edge, 2: assortative edge
edge_distribution = []
for i in G.edges():
    if G.node[i[0]]['timer'] != G.node[i[1]]['timer']:
        edge_distribution.append(1)
    else:
        edge_distribution.append(2)
plt.hist(edge_distribution, alpha = 0.5)
```

As we could see from the above results, we could see that ER graph with half of nodes with timer $0$ and the rest with timer $8$ will give you $0$ assortativity, which means the effect of assortativity is zero. It is natural as there is no relationship between the assignment of timer of a node and the timers of its neighbours.

### 1.2 Analytically calculate the expected assortativity based on the fraction of (dis)assortative edge fractions

Let's see if the obtained assortativity agrees with theoretical result. We use the equation introduced in [theoretical assortativity](http://arxiv.org/pdf/cond-mat/0209450.pdf):
$$r = \frac{\sum_i e_{i,i} - \sum_i a_ib_i}{1 - \sum_i a_ib_i}$$
where
$$a_i=\sum_j e_{i,j}$$

In order to analytically calculate the assortativity that matches the above result, we should set the fraction of assortative edges $0.5$ -- it will naturally set the fraction of disassortative edges $0.5$ as well.

```python
number_of_communities = 2.0 # Two communities
assortative_edge_fraction = 0.5/number_of_communities # e_i_i
disassortative_edge_fraction = (1/number_of_communities - assortative_edge_fraction)/(number_of_communities-1)
# e_i_j = (1/number_of_communities - assortative_edge_fraction)/(number_of_communities-1)
sum_e_i_i = number_of_communities*assortative_edge_fraction
sum_a_i_b_i = number_of_communities*(assortative_edge_fraction + (number_of_communities-1)*disassortative_edge_fraction)**2
# sum_a_i_b_i = 1
theoretical_assortativity = (sum_e_i_i - sum_a_i_b_i)/(1 - sum_a_i_b_i)
print theoretical_assortativity
```

# 2. Create a random graph with given assortativity

## 2.1. Fixed-point rewiring process
In this section, we will learn to create networks with ___given node attribute assortativity,___ based on 'timer' attribute. Rewire edges based on the algorithm introduced in [assortative mixing in networks](http://journals.aps.org/prl/abstract/10.1103/PhysRevLett.89.208701):

_First, we generate a random graph with the desired degree distribution according to the prescription given in Ref. [16]. Then we apply a Metropolis dynamics to the graph in which on each step we choose at random two edges, denoted by the vertex pairs, $(v_1, w_1)$ and $(v_2, w_2)$, that they connect. We measure the remaining degrees $(j_1,k_1)$ and $(j_2,k_2)$ for these vertex pairs, and then replace the edges with two new ones $(v_1,v_2)$ and $(w_1,w_2)$ __with probability $\min(1,\frac{e_{j_1,j_2}e_{k_1,k_2}}{e_{j_1,k_1}e_{j_2,k_2}})$.__ This dynamics conserves the degree sequence, is __ergodic__ on the set of graphs having that degree sequence, and, with the choice of acceptance probability above, satisfies detailed balance for state probabilities $\Pi_i e_{{j_i}{k_i}}$, and hence has the required edge distribution $e_{jk}$ as its __fixed point.__ _ <br />

$e_{jk}$ is calculated as the fraction edges in a network that connect a vertex of type $j$ to one of type $k$. (from [Mixing patterns in networks](http://journals.aps.org/pre/abstract/10.1103/PhysRevE.67.026126))
<br /><br />
__P.S There are many parts in this algorithm which is subtly made up in order to prevent the script from making mistakes. We'll go through the details through the following cell.__

#### 1) Create a random graph

```python
import networkx as nx
import numpy as np
import random as rd

network_Size = 10000
mean_Degree = 5.0
G = nx.fast_gnp_random_graph(network_Size, mean_Degree/network_Size)
G = sorted(nx.connected_component_subgraphs(G), key = len, reverse = True)[0]
```

In order to confirm if the degree distribution does not change after rewiring process, let's save the degree distribution of this graph in a variable __compare_degree_distribution__ and compare this later with the degree distribution of rewired-assortative network.

```python
compare_degree_distribution = G.degree().values()
```

#### 2) Assign timers as many as the number of communities you would like to create.
You should have nodal attribute that decides the membership of a node: the number of discrete values of nodal attribute will decide the number of different membership (communities). Here we'll divide a network into two groups: nodes with timer $0$ and the rest with timer $8$.
<br /><br />
Here, you should be careful __NOT__ to write as the following:

```python
number_of_Community = 2.0
for i in G:
    if i < network_Size/number_of_Community:
        G.node[i]['timer'] = 0
    else:
        G.node[i]['timer'] = 8
```

Above code will assign timer $\tau=0$ to low-number nodes and $\tau=8$ to high-number nodes. An edge $(x,y)$ of a network created by __Networkx__ is always ordered in a way $x<y$. Therefore, the pair of timers of an edge can always be:

$$(0,0), (0,8), (8,8)$$

even though the pair of timers should also include $(8,0)$, so

$$(0,0), (0,8), (8,0), (8,8).$$

This causes a problem when randomly rewiring edges. When rewiring edges, you choose two edges: $$[(x_1,y_1), (x_2,y_2)]$$
and with some probability, the two edges will either not rewired, or rewired to 
$$[(x_1,x_2), (y_1,y_2)].$$

The thing is, by networkx construction, $x_1$ and $x_2$ are strictly smaller than $y_1$ and $y_2$ respectively. Therefore, the above code will always give an edge where the timer of $x_1$ is always strictly smaller than that of $y_1$, which will only create 

$$(0,0), (0,8), (8,8) \rightarrow (0,0), (0,8), (8,8)$$

Therefore, we should allow the low-number nodes to have $\tau=8$ and high-number nodes to have $\tau=0$, so that the mixing of timer pairs can be:

$$(0,0), (0,8), (8,0), (8,8) \rightarrow (0,0), (0,8), (8,0), (8,8)$$

(we'll look into this more in depth later).
<br />
<br />
The correct way to assign timers is, therefore, is as the following:

```python
nodes_list = G.nodes()
rd.shuffle(nodes_list)
number_of_nodes_with_homogeneous_timers = 0
for i in nodes_list:
    if number_of_nodes_with_homogeneous_timers < network_Size/number_of_Community:
        G.node[i]['timer'] = 0
    else:
        G.node[i]['timer'] = 8
    number_of_nodes_with_homogeneous_timers += 1
print 'Assortativity of original network: ' + str(nx.attribute_assortativity_coefficient(G,'timer'))
```

You can also find that the timer-assortativity of ER network is close to zero, which is normal as timers are distributed randomly.

After assigning timers, now we need to decide the fraction $e_{ii}$ of assortative edges within a community (and automatically disassortative edge fraction) that we desire for a network to have. As we use undirected graph in this case, the mixing matrix is symmetric (i.e., $e_{ij}=e_{ji}$ for $i \neq j$), and also the edge densities of intra-community and inter-community are the same throughout all the communities (i.e., $e_{ii}=D$, $e_{ij}=d$ for all $i,j$). In short, in matrix form, the mixing matrix of an assortative network we are trying to make should be:

$$\textbf{e}= \begin{pmatrix}
D & d & ... & d & d\\ 
d & D & ... & d & d\\
... & ... & ... & ... & ...\\
d & d & ... & D & d\\
d & d & ... & d & D\\
\end{pmatrix}$$

Therefore, the total density of edges connected to community $i$ is $a_i = \sum_j e_{ij} = D + (C-1)d=\frac{1}{C}$ where C is the number of communities, and it is constant for all $i$. Therefore disassortative edge fraction is
$$d=\frac{\frac{1}{C}-D}{C-1}$$.

Therefore, the elements $e_{ij}$ of mixing matrix __e__ is:
$$
e_{ij}=\left\{
\begin{array}{l l}
D & \quad \text{if $i=j$,}\\
\frac{\frac{1}{C}-D}{C-1} & \quad \text{otherwise.}
\end{array} \right.
$$

And analytical value of assortativity is

$$\sum_i e_{ii} - \sum_i a_ib_i$$

Let's see if the assortativity we obtained after rewiring 100000 times matches with the [theoretical assortativity](http://arxiv.org/pdf/cond-mat/0209450.pdf):
$$r = \frac{\sum_i e_{ii} - \sum_i a_ib_i}{1 - \sum_i a_ib_i}$$
where
$$a_i=\sum_j e_{ij}\text{, }b_i=\sum_j e_{ji}.$$
and therefore
$$a_i=b_i=\frac{1}{C},$$
as that is what we assumed in the beginning (inter-and intra-edge densities and identical between among different communities). Therefore the assortativity $r$ is

$$r = \frac{CD - \frac{1}{C}}{\frac{C-1}{C}}$$

which gives

$$D = \frac{r-\frac{r-1}{C}}{C}$$
$$d = \frac{1-[r-(\frac{r-1}{C})]}{C(C-1)}$$

```python
assortativity = 0.8
assortative_edge_fraction = (assortativity - (assortativity-1)/number_of_Community)/number_of_Community
disassortative_edge_fraction = (1/number_of_Community - assortative_edge_fraction)/(number_of_Community-1)

print assortative_edge_fraction
print disassortative_edge_fraction
```

Based on the intra- and inter-edge densities calculated by assortativity and the number of communities, we can now generate an assortative network, getting rewired until the assortativity converges in the limit.

```python
past_average_assortativity = nx.attribute_assortativity_coefficient(G,'timer')
while 1:
    list_of_assortativity = []
    for i in xrange(1000):
        # Rewire until you find two distinctive edges that have no overlapping node
        while 1:
            random_edge_1 = G.edges()[rd.randint(0,G.size()-1)]
            random_edge_2 = G.edges()[rd.randint(0,G.size()-1)]
            if random_edge_1[0] not in G[random_edge_2[0]] and random_edge_1[0] not in G[random_edge_2[1]]:
                if random_edge_1[1] not in G[random_edge_2[0]] and random_edge_1[1] not in G[random_edge_2[1]]:
                    break

        timer_random_edge_1_node_1 = G.node[random_edge_1[0]]['timer']
        timer_random_edge_1_node_2 = G.node[random_edge_1[1]]['timer']
        timer_random_edge_2_node_1 = G.node[random_edge_2[0]]['timer']
        timer_random_edge_2_node_2 = G.node[random_edge_2[1]]['timer']
        
        
        if timer_random_edge_1_node_1 == timer_random_edge_1_node_2:
            frac_1_1_to_1_2 = assortative_edge_fraction
        else:
            frac_1_1_to_1_2 = disassortative_edge_fraction            
        if timer_random_edge_1_node_1 == timer_random_edge_2_node_1:
            frac_1_1_to_2_1 = assortative_edge_fraction
        else:
            frac_1_1_to_2_1 = disassortative_edge_fraction
        if timer_random_edge_2_node_2 == timer_random_edge_2_node_1:
            frac_2_2_to_2_1 = assortative_edge_fraction
        else:
            frac_2_2_to_2_1 = disassortative_edge_fraction
        if timer_random_edge_2_node_2 == timer_random_edge_1_node_2:
            frac_2_2_to_1_2 = assortative_edge_fraction
        else:
            frac_2_2_to_1_2 = disassortative_edge_fraction

            
        if rd.random() < (frac_1_1_to_2_1*frac_2_2_to_1_2)/(frac_1_1_to_1_2*frac_2_2_to_2_1):
            G.remove_edges_from([random_edge_1, random_edge_2])
            G.add_edges_from([(random_edge_1[0], random_edge_2[0]), (random_edge_1[1],random_edge_2[1])])
        
        list_of_assortativity.append(nx.attribute_assortativity_coefficient(G,'timer'))
    
    # Check until assortativity converges
    if abs(np.mean(list_of_assortativity) - past_average_assortativity) < 1e-2:
        # Once assortativity converges, check if assortativity of the last network is similar to assortativity.
        if abs(nx.attribute_assortativity_coefficient(G,'timer') - assortativity) < 1e-2:
            break
    past_average_assortativity = np.mean(list_of_assortativity)
    print past_average_assortativity
```

## 2.2. Configuration model with degree sequence

1. We first choose a size for our graph in terms of the
number M of edges and we draw M edges from
the distribution eij . That is, we generate M edges,
each one identified by the types of the vertices that
it connects, in some manner such that the fraction
of edges connecting vertices of types i and j tends
to eij as M becomes large. In practice, a simple
transformation method works well [36].

```python
e_i_i = 0.9
e_i_j = 1 - e_i_i

edge_set_size = 100 # M of edges from distribution e_ij
edge_set = []

for i in xrange(int(edge_set_size*e_i_i/2)):
    edge_set.append(['G1-', 'G1-'])
    edge_set.append(['G2-', 'G2-'])
for i in xrange(int(edge_set_size*e_i_j)):
    edge_set.append(['G1-', 'G2-'])
```

We count the number of ends of edges of each
type $i$, to give the sums $m_i$ of the degrees of vertices
in each class. We calculate the expected number $n_i$
of vertices of each type from ni = mi/zi (rounded
to the nearest integer), where zi
is the desired mean
degree of vertices of type i.

```python
sum_of_degrees_of_type_i = edge_set_size*e_i_i + edge_set_size*e_i_j # The number of edges of type i
sum_of_degrees_of_type_j = edge_set_size*e_i_i + edge_set_size*e_i_j # The number of edges of type i
mean_degree_of_type_i = 5
mean_degree_of_type_j = 5
number_of_type_i = sum_of_degrees_of_type_i/mean_degree_of_type_i
number_of_type_j = sum_of_degrees_of_type_j/mean_degree_of_type_j
```

 We draw ni vertices from the desired degree distribution
$p^{(i)}_k$
for type i. In general the degrees of
these vertices will not sum exactly to mi as we want
them to. So we choose one vertex at random, discard
it, and draw another from the distribution p
(i)
k
until the sum does equal mi
.

```python
def poisson_random_variable(mean_of_dist):
	random_number = rd.random()
	idx_of_rd_number = 0
	while 1:
		random_number *= rd.random()
		idx_of_rd_number += 1
		if random_number < math.exp(-mean_of_dist):
			return idx_of_rd_number
        
desired_degree_distribution_of_type_i = [poisson_random_variable(mean_degree_of_type_i) for i in range(int(number_of_type_i))]
while sum(desired_degree_distribution_of_type_i) != sum_of_degrees_of_type_i:
    desired_degree_distribution_of_type_i.pop(rd.randint(0, len(desired_degree_distribution_of_type_i)-1))
    desired_degree_distribution_of_type_i.append(poisson_random_variable(mean_degree_of_type_i))
    
node = 0
degree_sequence_i = []
for degree in desired_degree_distribution_of_type_i:
    for stub in range(degree):
        degree_sequence_i.append(node)
    node += 1
print degree_sequence_i
    
desired_degree_distribution_of_type_j = [poisson_random_variable(mean_degree_of_type_j) for i in range(int(number_of_type_j))]
while sum(desired_degree_distribution_of_type_j) != sum_of_degrees_of_type_j:
    desired_degree_distribution_of_type_j.pop(rd.randint(0, len(desired_degree_distribution_of_type_j)-1))
    desired_degree_distribution_of_type_j.append(poisson_random_variable(mean_degree_of_type_j))

node = 0
degree_sequence_j = []
for degree in desired_degree_distribution_of_type_j:
    for stub in range(degree):
        degree_sequence_j.append(node)
    node += 1
print degree_sequence_j
```

But are the two ways of doing assortative algorithms the same?

Kristina's way: Do exchange edges _until_ the desired assortativity is obtained.
